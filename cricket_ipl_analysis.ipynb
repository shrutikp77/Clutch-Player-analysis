##  Step 2: Context Features

## 2.1 ➤ runs_required, balls_remaining, wickets_remaining ##

```python
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
```

```python
matches.columns
```

```python
deliveries.columns
```

```python
matches.rename(columns={'id': 'match_id'}, inplace=True)
```

```python
# Merge target score from matches into deliveries
first_innings_scores = deliveries[deliveries['inning'] == 1].groupby('match_id')['total_runs'].sum().reset_index()
first_innings_scores.columns = ['match_id', 'target_score']
first_innings_scores['target_score'] += 1  # target = runs to win

# Merge target into deliveries
deliveries = deliveries.merge(first_innings_scores, on='match_id', how='left')

# Only keep second innings deliveries
second_innings = deliveries[deliveries['inning'] == 2].copy()

# Sort for cumulative features
second_innings.sort_values(by=['match_id', 'over', 'ball'], inplace=True)

# Cumulative runs
second_innings['cum_runs'] = second_innings.groupby('match_id')['total_runs'].cumsum()
second_innings['runs_required'] = second_innings['target_score'] - second_innings['cum_runs']

# Balls remaining
second_innings['ball_number'] = ((second_innings['over'] - 1) * 6 + second_innings['ball'])
second_innings['balls_remaining'] = 120 - second_innings['ball_number']  # 20 overs

# Wickets remaining
second_innings['player_dismissed_flag'] = second_innings['player_dismissed'].notnull().astype(int)
second_innings['cum_wickets'] = second_innings.groupby('match_id')['player_dismissed_flag'].cumsum()
second_innings['wickets_remaining'] = 10 - second_innings['cum_wickets']
```

```python
# Avoid division by zero
second_innings['current_RR'] = second_innings['cum_runs'] / (second_innings['ball_number'] / 6).replace(0, 0.1)
second_innings['required_RR'] = second_innings['runs_required'] / (second_innings['balls_remaining'] / 6).replace(0, 0.1)
```

```python
def get_phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Death'

second_innings['game_phase'] = second_innings['over'].apply(get_phase)
```

```python
X = 2.0  # sensitivity threshold
second_innings['is_pressure'] = (
    (second_innings['required_RR'] > second_innings['current_RR'] + X) |
    (second_innings['wickets_remaining'] <= 3)
).astype(int)
```

```python
# Per match batsman stats
bat_stats = deliveries.groupby(['match_id', 'batsman']).agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).reset_index()
bat_stats.rename(columns={'ball': 'balls_faced'}, inplace=True)

# Merge match date
bat_stats = bat_stats.merge(matches[['match_id', 'date']], on='match_id')
bat_stats['date'] = pd.to_datetime(bat_stats['date'])

# Sort for rolling
bat_stats.sort_values(['batsman', 'date'], inplace=True)

# Rolling strike rate
bat_stats['rolling_strike_rate'] = (
    bat_stats.groupby('batsman')
    .apply(lambda x: x['batsman_runs'].rolling(window=5, min_periods=1).sum() /
                     x['balls_faced'].rolling(window=5, min_periods=1).sum() * 100)
    .reset_index(drop=True)
)
```

```python
# Per match batsman stats
bat_stats = deliveries.groupby(['match_id', 'batsman']).agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).reset_index()
bat_stats.rename(columns={'ball': 'balls_faced'}, inplace=True)

# Merge match date
bat_stats = bat_stats.merge(matches[['match_id', 'date']], on='match_id')
bat_stats['date'] = pd.to_datetime(bat_stats['date'])

# Sort for rolling
bat_stats.sort_values(['batsman', 'date'], inplace=True)

# Rolling strike rate
bat_stats['rolling_runs'] = (
    bat_stats.groupby('batsman')['batsman_runs']
    .transform(lambda x: x.rolling(window=5, min_periods=1).sum())
)

bat_stats['rolling_balls'] = (
    bat_stats.groupby('batsman')['balls_faced']
    .transform(lambda x: x.rolling(window=5, min_periods=1).sum())
)

bat_stats['rolling_strike_rate'] = (bat_stats['rolling_runs'] / bat_stats['rolling_balls']) * 100
```

```python
bat_stats['form_index'] = (
    bat_stats.groupby('batsman')['batsman_runs']
    .rolling(window=5, min_periods=1)
    .mean()
    .reset_index(drop=True)
)
```

```python
# Calculate runs scored under pressure for each batsman
clutch_stats = second_innings.groupby(['batsman', 'is_pressure']).agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).reset_index()

# Pivot: pressure vs non-pressure
clutch_pivot = clutch_stats.pivot(index='batsman', columns='is_pressure', values='batsman_runs').fillna(0)
clutch_pivot.columns = ['non_pressure_runs', 'pressure_runs']
clutch_pivot['clutch_score'] = clutch_pivot['pressure_runs'] / (clutch_pivot['non_pressure_runs'] + 1)
```

```python
clutch_pivot['ppi'] = (
    0.5 * clutch_pivot['clutch_score'] +
    0.5 * bat_stats.groupby('batsman')['form_index'].mean()
)
```

```python
import matplotlib.pyplot as plt

def plot_win_probability_curve(df, match_id):
    match_df = df[df['match_id'] == match_id].copy()
    plt.figure(figsize=(12, 5))
    plt.plot(match_df['ball_number'], match_df['win_prob'], label='Win Probability')
    plt.axhline(0.5, color='gray', linestyle='--')
    plt.xlabel('Ball Number')
    plt.ylabel('Win Probability')
    plt.title(f'Win Probability Over Time - Match {match_id}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Create win labels: 1 if batting_team won in 2nd innings
match_winners = matches[['match_id', 'winner']]

# Merge into second_innings
second_innings = second_innings.merge(match_winners, on='match_id', how='left')
second_innings['win'] = (second_innings['batting_team'] == second_innings['winner']).astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

features = ['runs_required', 'balls_remaining', 'wickets_remaining', 'current_RR', 'required_RR']
X = second_innings[features]
y = second_innings['win']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

model = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=42)
model.fit(X_train, y_train)

y_pred_prob = model.predict_proba(X_test)[:, 1]
print("AUC Score:", roc_auc_score(y_test, y_pred_prob))
```

```python
second_innings['win_prob'] = model.predict_proba(second_innings[features])[:, 1]
```

```python
second_innings['prev_win_prob'] = second_innings.groupby('match_id')['win_prob'].shift(1)
second_innings['wp_delta'] = second_innings['win_prob'] - second_innings['prev_win_prob']

# Clutch Impact Per Player
clutch_impact = second_innings.groupby('batsman')['wp_delta'].mean().reset_index()
clutch_impact.columns = ['batsman', 'avg_wp_delta']
clutch_impact.sort_values('avg_wp_delta', ascending=False).head(10)
```
